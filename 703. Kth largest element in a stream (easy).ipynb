```python
class MinHeap:
    def __init__(self):
        self.heap = []

    def push(self,val):
        self.heap.append(val)
        curr = len(self.heap) - 1
        if curr == 0:
            return

        parent = (curr - 1) // 2

        while curr > 0 and self.heap[curr] < self.heap[parent]:
            self.heap[curr], self.heap[parent] = self.heap[parent], self.heap[curr]
            curr = parent
            parent = (curr - 1) // 2
    
    def pop(self):
        if len(self.heap) == 0:
            return None
        if len(self.heap) == 1:
            return self.heap.pop()

        self.heap[0], self.heap[-1] = self.heap[-1], self.heap[0]
        val = self.heap.pop()
        curr = 0
        child_1 = curr*2 + 1
        child_2 = curr*2 + 2
        while child_1 < len(self.heap):
            if child_2 < len(self.heap) and self.heap[child_1]>self.heap[child_2]:
                min_child = child_2
            else:
                min_child = child_1
            if self.heap[curr] <= self.heap[min_child]:
                break
            else:    
                self.heap[curr], self.heap[min_child] = self.heap[min_child], self.heap[curr]
                curr = min_child
                child_1 = curr*2 + 1
                child_2 = curr*2 + 2

        return val

    def __getitem__(self,index):
        return self.heap[index]
    
    def __str__(self):
        return str(self.heap)

    def __len__(self):
        return len(self.heap)
```
